### Matrix als $\LaTeX$-Tabelle

```julia
for x in 1:10
    print(string(" & \$", x, "\$"))
end
println(" \\\\")

for y in 1:10
    print(string("\$", y, "\$ "))
    for x in 0:9
        pos = 10*x + y
        num = data[pos] / 1e6
        num = Int(round(num))

        print(string(" & \$", num, "\$"))
    end
    println(" \\\\")
end
```

## Heatmap

```julia
thermal_splitted = palette(:thermal, 10)
thermal_splitted
```

```julia
colorscale = cgrad([palette(:thermal, 10)[2], palette(:thermal, 10)[10]], scale=:log)
```

```julia
img = heatmap(
    data/1e6,
    c=colorscale,
    size=(600, 600)
)
title!(L"\mathrm{Heatmap\ der\ Zhlraten\ in\ 10^6\,Hz}")
xlabel!(L"x-\mathrm{Richtung}")
ylabel!(L"y-\mathrm{Richtung}")
xticks!(1:10)
yticks!(1:10)
```
